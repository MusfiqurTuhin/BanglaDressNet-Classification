```python


# Block 1: Google Drive Authentication and Dataset Unzipping
from google.colab import drive
import os
import matplotlib.pyplot as plt
import numpy as np

# Mount Google Drive to access the dataset
print("Mounting Google Drive...")
drive.mount('/content/gdrive')
print("Google Drive mounted.")

# Define the full path to the zip file in Google Drive
zip_file_path = '/content/gdrive/My Drive/BanglaDressNet_Dataset.zip'  # Adjust path if it's in a folder

unzip_destination = '/content/'
dataset_final_path = os.path.join(unzip_destination, 'BanglaDressNet_Dataset')

# Check and unzip
if not os.path.exists(zip_file_path):
    print(f"Error: Zip file not found at {zip_file_path}. Please check the path.")
else:
    print(f"Unzipping {zip_file_path} to {unzip_destination}...")
    # The -q flag is for quiet mode (suppress output)
    # The -d flag specifies the destination directory
    !unzip -q "{zip_file_path}" -d "{unzip_destination}"
    if os.path.exists(dataset_final_path):
        print(f"Contents of {dataset_final_path}:")
        # List contents of the unzipped directory for verification (uncomment for debugging)
        # !ls "{dataset_final_path}"
        print("Dataset unzipped successfully.")
    else:
        print(f"Error: Dataset directory not found at {dataset_final_path} after unzipping.")

# Block 2: Importing Libraries and TensorFlow Version Check
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
# Important: No Rescaling layer needed here if preprocess_input is inside the model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score, roc_auc_score, roc_curve, auc
import pandas as pd
import seaborn as sns

print(f"TensorFlow Version: {tf.__version__}")

# Block 3: Define Dataset Directories and Image Parameters
train_dir = os.path.join(dataset_final_path, 'train')
validation_dir = os.path.join(dataset_final_path, 'val')
test_dir = os.path.join(dataset_final_path, 'test')

# Define batch size and image dimensions for the model input
BATCH_SIZE = 32
IMG_SIZE = (224, 224)  # Input size

# Block 4: Load Datasets using image_dataset_from_directory
# tf.keras.utils.image_dataset_from_directory is used to load images from disk
# into a tf.data.Dataset.
# Images will be loaded as float32 in [0, 255] range.
print(f"\nLoading datasets from: {dataset_final_path}")
train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    shuffle=True,
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE,
    interpolation='bilinear' # Explicitly set interpolation method
)
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    validation_dir,
    shuffle=True,
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE,
    interpolation='bilinear'
)
test_dataset = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    shuffle=False, # Keep shuffle=False for test_dataset for consistent evaluation order
    batch_size=BATCH_SIZE,
    image_size=IMG_SIZE,
    interpolation='bilinear'
)

# Get class names from the training dataset
class_names = train_dataset.class_names
nb_classes = len(class_names)
print(f"Detected {nb_classes} classes: {class_names}")

# Configure datasets for performance
# .cache() keeps images in memory after they're loaded off disk during the first epoch.
# .prefetch() overlaps data preprocessing and model execution while training.
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.cache().prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.cache().prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.cache().prefetch(buffer_size=AUTOTUNE)

# Block 5: Data Augmentation Layer
# This sequential model applies various random transformations to images during training.
# These augmentations help the model generalize better to unseen data and reduce overfitting.
# This layer operates on images in the [0, 255] pixel range.
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal", name="random_flip"), # Randomly flip images horizontally
    layers.RandomRotation(0.1, name="random_rotation"), # Randomly rotate images by up to 10%
    layers.RandomZoom(0.1, name="random_zoom"), # Randomly zoom into images by up to 10%
    layers.RandomContrast(0.2, name="random_contrast"), # Randomly adjust contrast by up to 20%
    layers.RandomBrightness(0.2, name="random_brightness"), # Randomly adjust brightness by up to 20%
], name="data_augmentation_layer")

# Block 6: Visualize Augmented Images
# Display a batch of images with applied augmentations to visually confirm their effect.
plt.figure(figsize=(12, 12))
for images, labels in train_dataset.take(1): # Take one batch from the training dataset (images are [0, 255] float32)
    first_image = images[0] # Select the first image from the batch
    for i in range(9): # Display 9 augmented versions of this image
        ax = plt.subplot(3, 3, i + 1)
        # Apply data augmentation to the single image (expanded to batch dim)
        augmented_image = data_augmentation(tf.expand_dims(first_image, 0), training=True)
        # Display the augmented image. Convert to uint8 for proper display.
        plt.imshow(augmented_image[0].numpy().astype("uint8"))
        plt.axis("off") # Hide axes
plt.suptitle("Augmented Images (Example - pixel values [0, 255])", fontsize=16)
plt.show()

# Block 7: Load Base Model
# 'imagenet' weights are used for transfer learning.
# include_top=False means we don't include the classification head of the original model,
# allowing us to add our own custom classification layers.
base_model = keras.applications.ResNet50V2(
    weights="imagenet",
    input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3), # Input shape (height, width, channels)
    include_top=False, # Exclude the top (classification) layer
)

# Block 8: Freeze Base Model Layers
# Set the base model to non-trainable during the initial training phase.
# This prevents the pre-trained weights from being updated, focusing training
# on the newly added classification layers.
base_model.trainable = False

# Block 9: Display Base Model Summary
# Print a summary of the base model's architecture.
base_model.summary()

# Block 10: Build the Transfer Learning Model
# Define the input layer for the model.
inputs = keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3), name="input_image")
# Apply data augmentation to the input images. These are initially [0, 255] float32.
x = data_augmentation(inputs)
# Apply Xception's specific preprocessing. This function transforms [0, 255] to the expected range.
# Corrected line: Call preprocess_input from the xception module
x = tf.keras.applications.resnet_v2.preprocess_input(x)
# Pass the preprocessed images through the frozen base model.
# training=False ensures batch normalization layers run in inference mode during this phase.
x = base_model(x, training=False)
# Apply Global Average Pooling to reduce spatial dimensions to a single feature vector.
x = GlobalAveragePooling2D(name="global_average_pooling_layer")(x)
# Add a Dropout layer to prevent overfitting. Increased from 0.2 to 0.3 for better regularization.
x = Dropout(0.3, name="dropout_layer")(x)
# Add a dense layer with ReLU activation for feature learning.
x = Dense(512, activation="relu", name="dense_512_layer")(x)
# Add the final dense output layer with softmax activation for multi-class classification.
outputs = Dense(nb_classes, activation="softmax", name="output_layer")(x)
# Create the complete transfer learning model.
model = keras.Model(inputs, outputs, name="transfer_learning_model")

# Block 11: Display Full Model Summary
# Print a summary of the complete model's architecture, including the new layers.
model.summary()

# Block 12: Define Callbacks (ReduceLROnPlateau, EarlyStopping)
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping

# ReduceLROnPlateau: Reduces the learning rate when a metric has stopped improving.
# monitor='val_loss': Monitor validation loss.
# factor=0.2: New learning rate will be old_lr * 0.2. (Changed from 0.5 for more aggressive reduction)
# patience=5: Number of epochs with no improvement after which learning rate will be reduced. (Increased from 3)
# min_lr=1e-7: Lower bound on the learning rate. (Changed from 1e-6)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-7, verbose=1)

# EarlyStopping: Stops training when a monitored metric has stopped improving.
# monitor='val_loss': Monitor validation loss.
# patience=15: Number of epochs with no improvement after which training will be stopped. (Increased from 10)
# restore_best_weights=True: Restores model weights from the epoch with the best monitored value.
early_stopping = EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True, verbose=1)

# Block 13: Compile the Model (Initial Training)
# Compile the model with an optimizer, loss function, and metrics.
# 'adam' optimizer is a good general-purpose optimizer.
# 'sparse_categorical_crossentropy' is used for integer labels.
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Block 14: Fit the Model (Initial Training)
# Train the model on the training dataset and validate on the validation dataset.
# The `callbacks` list includes the defined ReduceLROnPlateau and EarlyStopping callbacks.
print("\nStarting initial training (feature extraction)...")
ehistory = model.fit(
    train_dataset,
    epochs=20, # Number of epochs for initial training
    validation_data=validation_dataset,
    callbacks=[reduce_lr, early_stopping]
)
print("Initial training complete.")

# Block 15: Evaluate Model on Test Set (After Initial Training)
# Evaluate the model's performance on the unseen test dataset after initial training.
print("\nEvaluating model on test set after initial training...")
scores_initial = model.evaluate(test_dataset)
print(f"Initial Test Accuracy: {scores_initial[1]*100:.2f}%")

# Block 16: Unfreeze Base Model for Fine-tuning
# Set the base model to trainable to allow fine-tuning of its weights.
# This is done after initial training of the new top layers,
# allowing the pre-trained features to be slightly adjusted for the specific task.
base_model.trainable = True
print("\nBase model unfrozen for fine-tuning.")

# Block 17: Compile the Model (Fine-tuning)
# Recompile the model with a very low learning rate for fine-tuning.
# A small learning rate prevents large updates to the pre-trained weights,
# preserving the learned features while adapting them.
model.compile(
    optimizer=keras.optimizers.Adam(1e-5), # Very low learning rate for fine-tuning
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
print("Model recompiled for fine-tuning.")

# Block 18: Fit the Model (Fine-tuning)
# Continue training the model, now with the base model layers unfrozen.
# Increased epochs to 100, but EarlyStopping will prevent overfitting by stopping early if needed.
print("\nStarting fine-tuning...")
ehistory_ft = model.fit(
    train_dataset,
    epochs=100, # Increased epochs for fine-tuning
    validation_data=validation_dataset,
    callbacks=[reduce_lr, early_stopping]
)
print("Fine-tuning complete.")

# Block 19: Convert Model to TFLite
# Convert the trained Keras model to TensorFlow Lite format for deployment on edge devices.
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
with open("model.tflite", 'wb') as f:
    f.write(tflite_model)
print("Model converted to TFLite and saved as model.tflite.")

# Block 20: Generate Evaluation Plots for Paper
# This block generates various evaluation metrics and plots to assess the model's performance comprehensively.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    roc_curve,
    auc
)
import tensorflow as tf

##############################
# Train, Val, Test Accuracy  #
##############################
print("\n==== Train, Validation, and Test Accuracy ====")
# Extract final train and val accuracy from training history
final_train_acc = ehistory_ft.history['accuracy'][-1]
final_val_acc = ehistory_ft.history['val_accuracy'][-1]

# Evaluate on test set
test_loss, test_acc = model.evaluate(test_dataset, verbose=0)
print(f"Train Accuracy         : {final_train_acc:.4f}")
print(f"Validation Accuracy    : {final_val_acc:.4f}")
print(f"Test Accuracy          : {test_acc:.4f}")

# Save accuracy scores to CSV
overall_score_df = pd.DataFrame({
    'Metric': ['Train Accuracy', 'Validation Accuracy', 'Test Accuracy'],
    'Score': [final_train_acc, final_val_acc, test_acc]
})
overall_score_df.to_csv("overall_scores.csv", index=False)
print("Overall scores saved to overall_scores.csv")

##########################
# Get Predictions        #
##########################
print("\nGenerating predictions for classification report and metrics...")
true_labels = []
pred_labels = []
pred_probs = []

for images, labels in test_dataset:
    probs = model.predict(images, verbose=0)
    preds = np.argmax(probs, axis=1)
    true_labels.extend(labels.numpy())
    pred_labels.extend(preds)
    pred_probs.extend(probs)

true_labels = np.array(true_labels)
pred_labels = np.array(pred_labels)
pred_probs = np.array(pred_probs)

##########################
# Classification Report  #
##########################
report = classification_report(true_labels, pred_labels, target_names=class_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print("\nClassification Report:\n", report_df)

# Save classification report to CSV
report_df.to_csv("classification_report.csv")
print("Classification report saved to classification_report.csv")

# Extract per-class metrics
precision_per_class = report_df['precision'].drop(['accuracy', 'macro avg', 'weighted avg']).values
recall_per_class = report_df['recall'].drop(['accuracy', 'macro avg', 'weighted avg']).values
f1_per_class = report_df['f1-score'].drop(['accuracy', 'macro avg', 'weighted avg']).values

# Save per-class metrics to CSV
metrics_df = pd.DataFrame({
    'Class': class_names,
    'Precision': precision_per_class,
    'Recall': recall_per_class,
    'F1 Score': f1_per_class
})
metrics_df.to_csv("classwise_metrics.csv", index=False)
print("Class-wise metrics saved to classwise_metrics.csv")

##############################
# Overall Metrics (Text)     #
##############################
overall_accuracy = accuracy_score(true_labels, pred_labels)
overall_precision = report_df.loc['weighted avg', 'precision']
overall_recall = report_df.loc['weighted avg', 'recall']
overall_f1 = report_df.loc['weighted avg', 'f1-score']
roc_auc = roc_auc_score(true_labels, pred_probs, multi_class='ovo', average='weighted')

print("\nOverall Scores:")
print(f"Accuracy: {overall_accuracy:.4f}")
print(f"Precision: {overall_precision:.4f}")
print(f"Recall: {overall_recall:.4f}")
print(f"F1 Score: {overall_f1:.4f}")
print(f"ROC-AUC Score: {roc_auc:.4f}")

#####################
# Confusion Matrix  #
#####################
cm = confusion_matrix(true_labels, pred_labels)
plt.figure(figsize=(12, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.savefig("confusion_matrix.png")
plt.show()
print("Confusion matrix plot saved as confusion_matrix.png")

##################
# ROC Curve Plot #
##################
plt.figure(figsize=(10, 8))
for i in range(len(class_names)):
    fpr, tpr, _ = roc_curve(true_labels == i, pred_probs[:, i])
    plt.plot(fpr, tpr, label=f"{class_names[i]} (AUC={auc(fpr, tpr):.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve by Class")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.savefig("roc_curve.png")
plt.show()
print("ROC curve plot saved as roc_curve.png")

##########################
# Accuracy and Loss Plot #
##########################
plt.figure(figsize=(8, 5))
plt.plot(ehistory_ft.history['accuracy'], label='Train Accuracy')
plt.plot(ehistory_ft.history['val_accuracy'], label='Validation Accuracy')
plt.title("Training and Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.savefig("accuracy_curve.png")
plt.show()
print("Accuracy curve plot saved as accuracy_curve.png")

plt.figure(figsize=(8, 5))
plt.plot(ehistory_ft.history['loss'], label='Train Loss')
plt.plot(ehistory_ft.history['val_loss'], label='Validation Loss')
plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig("loss_curve.png")
plt.show()
print("Loss curve plot saved as loss_curve.png")

# Block 21: Sample Predictions with Confidence Scores
# This block randomly selects images from the test set, makes predictions,
# and prints the true label, predicted label, and confidence score.
import random
import tensorflow as tf
import numpy as np

# Parameters for sample prediction
num_samples = 20 # Increased for better visual insight
IMG_SIZE = (224, 224)
test_dir_for_samples = os.path.join(dataset_final_path, 'test')

assert 'class_names' in globals(), "class_names not defined. Please ensure previous blocks are run."

# Gather all image paths and their corresponding true labels from the test directory
all_image_paths = []
all_labels = []
for class_name_idx, class_name in enumerate(class_names):
    class_path = os.path.join(test_dir_for_samples, class_name)
    if os.path.isdir(class_path):
        for img_name in os.listdir(class_path):
            if img_name.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff')):
                all_image_paths.append(os.path.join(class_path, img_name))
                # Map true label name to its index in the `class_names` list
                try:
                    all_labels.append(class_names.index(class_name))
                except ValueError:
                    print(f"Warning: Class '{class_name}' not found in model's class_names. Skipping image {img_name}")

# Define the preprocessing function for individual images.
# This function should load, resize, and cast to float32, keeping pixels in [0, 255].
def preprocess_for_model(img_path):
    img_raw = tf.io.read_file(img_path)
    img_tensor = tf.image.decode_image(img_raw, channels=3, expand_animations=False)
    img_tensor = tf.image.resize(img_tensor, IMG_SIZE, method=tf.image.ResizeMethod.BILINEAR)
    img_tensor = tf.cast(img_tensor, tf.float32) # Convert to float32, keep in [0, 255] range
    return tf.expand_dims(img_tensor, axis=0), img_tensor # Return batched tensor for model, and unbatched for display


# Select a random subset of images for prediction
if len(all_image_paths) >= num_samples:
    indices = random.sample(range(len(all_image_paths)), num_samples)
    sample_paths = [all_image_paths[i] for i in indices]
    sample_labels = [all_labels[i] for i in indices]

    print("\n===== Sample Predictions with Confidence =====")
    plt.figure(figsize=(20, int(np.ceil(num_samples / 5)) * 5)) # Adjust figure size dynamically

    for i, (img_path, true_idx) in enumerate(zip(sample_paths, sample_labels)):
        # Preprocess the image for model input
        img_array_processed, original_img_tensor_for_display = preprocess_for_model(img_path)

        # Make a single prediction using the trained model
        pred_probs = model.predict(img_array_processed, verbose=0)
        pred_idx = np.argmax(pred_probs)
        confidence = pred_probs[0][pred_idx]

        true_label = class_names[true_idx]
        pred_label = class_names[pred_idx]
        is_correct = (true_idx == pred_idx)
        color = "green" if is_correct else "red"

        print(f"[{i+1}] {os.path.basename(img_path)}")
        print(f"    True Label    : {true_label}")
        print(f"    Predicted     : {pred_label} ({'Correct' if is_correct else 'INCORRECT'})")
        print(f"    Confidence    : {confidence:.4f}")
        print("")

        # original_img_tensor_for_display is in [0, 255] float32, cast to uint8 for display
        display_img = original_img_tensor_for_display.numpy().astype("uint8")

        plt.subplot(int(np.ceil(num_samples / 5)), 5, i + 1)
        plt.imshow(display_img)
        plt.axis('off')
        plt.title(f"T: {true_label}\nP: {pred_label}\nConf: {confidence:.2f}", color=color)

    plt.tight_layout()
    plt.suptitle("Sample Predictions (Green: Correct, Red: Incorrect)", fontsize=16, y=1.02)
    plt.show()
else:
    print(f"Only found {len(all_image_paths)} test images. Need at least {num_samples} for sampling.")
```
